# single variable house price prediction linear regression

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/kaggle/input/simple-house-price-prediction/house_price.csv")
df.head()
```

remove rooms because we are practicing single variable 

```python
newDf = df.drop('rooms', axis=1)
newDf.head()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['area'], df['price'], color='blue', alpha=0.5, label='Actual Data')
plt.title('House Price vs. Square Footage', fontsize=16)
plt.xlabel('Square Footage (sqft)', fontsize=14)
plt.ylabel('Price ($)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
# Prepare data
X = df[['area']]
y = df['price']

# Train model
model = linear_model.LinearRegression()
model.fit(X,y)
```

```python
y_pred = model.predict(X)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['area'], df['price'], color='blue', alpha=0.5, label='Actual Data')
plt.plot(df['area'], y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('House Price vs. Square Footage (with Regression Line)', fontsize=16)
plt.xlabel('Square Footage (sqft)', fontsize=14)
plt.ylabel('Price ($)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
